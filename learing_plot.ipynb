```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
learn_log = pd.read_csv('catboost_info/learn_error.tsv', sep='\t')
test_log = pd.read_csv('catboost_info/test_error.tsv', sep='\t')
```

```python
learn_log.head()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(learn_log['iter'], learn_log.iloc[:, 1], label='Training Error')
plt.plot(test_log['iter'], test_log.iloc[:, 1], label='Validation Error')
plt.xlabel('Iterations')
plt.ylabel('Error')
plt.title('CatBoost Learning Curve')
plt.legend()
plt.show()
```
